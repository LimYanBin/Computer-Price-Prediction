## 1.3 Feature Engineering

### Feature Engineering 1
1. In here we create a new feature call Price(RM) that is converted from the feature Price (in Indian Rupees), the reason:
    - The trained model will be used in Malaysia, so that we need to provide our client the Price of laptop in RM.
2. The old feature will be drop after converted

```python
data['Price(RM)'] = (data['Price (in Indian Rupees)'] * 0.055).round(2)
data.drop(columns=['Price (in Indian Rupees)'], inplace=True)
```

### Feature Engineering 2
1. In here we gonna convert all the records in Refresh Rate into 1 for 60 refresh rate and 2 for others
2. The reason why we doing this is because the minimum refresh rate for current laptop is 60 but from the results that we discover, the refresh rate consists of 1, 2, 3, 6, 9
3. The old feature will be drop after converted

```python
data['Refresh Rate'].value_counts()
```

```python
def convert_refresh_rate(int):
    if int == 60:
        return 1
    else:
        return 2
    
data['Refresh_Rate'] = data['Refresh Rate'].apply(convert_refresh_rate)
data.drop(columns=['Refresh Rate'], inplace=True)
```
